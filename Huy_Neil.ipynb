```python
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np

from random import random
from numpy import log
from numpy import math
from scipy import misc


#----------------------------------------------------------------------------------------#
# Function

def fonction(x):
     return (x**4-16*(x**2)+5*x)/100
    # return      x*sin(4*pi*(abs(x+1)))
#----------------------------------------------------------------------------------------#
# Plot function


x = np.arange(-5.0, 5.0, 0.01)
# x = np.arange(0, 2.0, 0.01)
y = fonction(x)

plt.plot(x, y,'r-')
plt.show()

#----------------------------------------------------------------------------------------#
# Gradient Descent

error = [];
iters = [];
steps = [];


alpha = 0.5 # learning rate stepsize
beta = 0.5 # learning rate stepsize
nb_max_iter = 1000 # Nb max d'iteration
eps = 0.00001 # stop condition

x0 = 3 # start point
# x0 = 1.2
y0 = fonction(x0)
plt.scatter(x0, fonction(x0))

cond = eps + 10.0 # start with cond greater than eps (assumption)
nb_iter = 0 
tmp_y = y0
while cond > eps and nb_iter < nb_max_iter:
    x0 = x0 - (nb_iter+1)**(-alpha) * misc.derivative(fonction, x0)+math.sqrt((nb_iter+1)**(-beta))*math.sqrt(2)*math.sqrt((nb_iter+1)**(-alpha))*np.random.normal()/100
    y0 = fonction(x0)
    nb_iter = nb_iter + 1

    cond = abs( tmp_y - y0 )
    tmp_y = y0
    steps.append((nb_iter+1)**(-alpha))
    error.append(cond)
    iters.append(nb_iter)
    print (nb_iter,x0,y0,cond) 
    plt.scatter(x0, y0)
    

    

plt.title("Gradient Descent for quartic function")
plt.grid()

plt.savefig("gradient_descent_1d_python.png", bbox_inches='tight')
plt.show()
```
